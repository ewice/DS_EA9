```python
import numpy as np
import matplotlib.pyplot as plt

X = [6,8,4,1,6,4,3,8,5,2,4,8,6,2,3,6,3,8,5,2]
y = [0.53,0.8,0.39,0.2,0.33,0.16,0.53,0.08,.005,.12,0.24,1,0.39,0.2,0.33,0.16,0.53,0.08,.005,.12]

plt.scatter(X, y);

X_b = np.c_[np.ones((20, 1)), X]  
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best

X_new = np.array([[0], [20]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict 

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 12, 0, 1])
plt.show()
```
